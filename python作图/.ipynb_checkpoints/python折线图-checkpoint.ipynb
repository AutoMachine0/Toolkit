```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import MultipleLocator

# 字体设置
font1 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 15}

font2 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 12}

plt.subplot(2, 2, 1)

x = np.array([0.1, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])
y = np.array([0.912, 0.902, 0.916, 0.912, 0.932, 0.934, 0.875, 0.916, 0.875, 0.922, 0.927])
std_dev = np.array([0.015, 0.028, 0.008, 0.007, 0.013, 0.021, 0.011, 0.020, 0.005, 0.021, 0.005])

# 坐标轴间隔设置，ｘ间隔为100,y将为0.005
x_axis_space = MultipleLocator(0.5)
y_axis_space = MultipleLocator(0.02)

# 获取坐标轴对象
axis = plt.gca()
# 使用坐标轴实例设置坐标轴间隔
axis.xaxis.set_major_locator(x_axis_space)
axis.yaxis.set_major_locator(y_axis_space)

# 设置x轴刻度显示
value = ["0.1", "0.5", "1.0", "1.5", "2.0", "2.5", "3.0", "3.5", "4.0", "4.5", "5.0"]
plt.xticks(x, value)

# 开启网格
plt.grid(linestyle="--")

# 限制坐标轴显示范围
plt.xlim(0, 5.0)
plt.ylim(0.81, 0.96)

plt.tick_params(labelsize=13)
labels = axis.get_xticklabels() + axis.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]

# 创建折线图
plt.plot(x, y, marker='*', color="firebrick", label='Performance Avg')

# 创建标准差区间
plt.fill_between(x, y - std_dev, y + std_dev, color='salmon', alpha=0.2, label='Performance Std')

# 添加标签和标题c
plt.xlabel("(a) Micro Temperature Coefficient "+r'$\tau_{mic}$', fontdict=font1)
plt.ylabel('Average Validation Accuracy', fontdict=font1)
plt.legend(loc=3, prop=font2)

# 图2
plt.subplot(2, 2, 2)

x = np.array([50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550])
y = np.array([0.898, 0.925, 0.912, 0.89, 0.88, 0.86, 0.88, 0.89, 0.89, 0.872, 0.896])
std_dev = np.array([0.007, 0.015, 0.007, 0.020, 0.026, 0.015, 0.012, 0.024, 0.010, 0.016, 0.014])

# 坐标轴间隔设置，ｘ间隔为100,y将为0.005
x_axis_space = MultipleLocator(50)
y_axis_space = MultipleLocator(0.02)

# 获取坐标轴对象
axis = plt.gca()
# 使用坐标轴实例设置坐标轴间隔
axis.xaxis.set_major_locator(x_axis_space)
axis.yaxis.set_major_locator(y_axis_space)

# 设置x轴刻度显示
value = ["50", "100", "150", "200", "250", "300", "350", "400", "450", "500", "550"]
plt.xticks(x, value)

# 开启网格
plt.grid(linestyle="--")

# 限制坐标轴显示范围
plt.xlim(50, 550)
plt.ylim(0.81, 0.95)

plt.tick_params(labelsize=13)
labels = axis.get_xticklabels() + axis.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]

# 创建折线图
plt.plot(x, y, marker='o', color='dodgerblue', label='Performance Avg')

# 创建标准差区间
plt.fill_between(x, y - std_dev, y + std_dev, color='deepskyblue', alpha=0.2, label='Performance Std')

# 添加标签和标题c
plt.xlabel('(b) Micro Search Epoch '+r'$E_{mic}$', fontdict=font1)
plt.ylabel('Average Validation Accuracy', fontdict=font1)
plt.legend(loc=3, prop=font2)

# 图3
plt.subplot(2, 2, 3)

# macro search temperature=2.0
x = np.array([0.1, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])
y = np.array([0.862, 0.894, 0.87, 0.89, 0.933, 0.922, 0.863, 0.914, 0.894, 0.918, 0.910])
std_dev = np.array([0.015, 0.035, 0.010, 0.014, 0.017, 0.010, 0.033, 0.015, 0.022, 0.007, 0.024])

# 坐标轴间隔设置，ｘ间隔为100,y将为0.005
x_axis_space = MultipleLocator(0.5)
y_axis_space = MultipleLocator(0.02)

# 获取坐标轴对象
axis = plt.gca()
# 使用坐标轴实例设置坐标轴间隔
axis.xaxis.set_major_locator(x_axis_space)
axis.yaxis.set_major_locator(y_axis_space)

# 设置x轴刻度显示
value = ["0.1", "0.5", "1.0", "1.5", "2.0", "2.5", "3.0", "3.5", "4.0", "4.5", "5.0"]
plt.xticks(x, value)

# 开启网格
plt.grid(linestyle="--")

# 限制坐标轴显示范围
plt.xlim(0, 5.0)
plt.ylim(0.81, 0.955)

plt.tick_params(labelsize=13)
labels = axis.get_xticklabels() + axis.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]

# 创建折线图
plt.plot(x, y, marker='^', color="orange", label='Performance Avg')

# 创建标准差区间
plt.fill_between(x, y - std_dev, y + std_dev, color='gold', alpha=0.2, label='Performance Std')

plt.xlabel('(c) Macro Temperature Coefficient ' + r'$\tau_{mac}$', fontdict=font1)
plt.ylabel('Average Validation Accuracy', fontdict=font1)
plt.legend(loc=3, prop=font2)

# 图4
plt.subplot(2, 2, 4)

# macro search epoch=350
x = np.array([50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550])
y = np.array([0.918, 0.89, 0.87, 0.92, 0.882, 0.898, 0.924, 0.908, 0.917, 0.918, 0.90])
std_dev = np.array([0.015, 0.017, 0.030, 0.010, 0.015, 0.023, 0.01, 0.015, 0.005, 0.012, 0.015])

# 坐标轴间隔设置，ｘ间隔为100,y将为0.005
x_axis_space = MultipleLocator(50)
y_axis_space = MultipleLocator(0.02)

# 获取坐标轴对象
axis = plt.gca()
# 使用坐标轴实例设置坐标轴间隔
axis.xaxis.set_major_locator(x_axis_space)
axis.yaxis.set_major_locator(y_axis_space)

# 设置x轴刻度显示
value = ["50", "100", "150", "200", "250", "300", "350", "400", "450", "500", "550"]
plt.xticks(x, value)

# 开启网格
plt.grid(linestyle="--")

# 限制坐标轴显示范围
plt.xlim(50, 550)
plt.ylim(0.81, 0.95)

plt.tick_params(labelsize=13)
labels = axis.get_xticklabels() + axis.get_yticklabels()
[label.set_fontname('Times New Roman') for label in labels]

# 创建折线图
plt.plot(x, y, marker='v', color="purple", label='Performance Avg')

# 创建标准差区间
plt.fill_between(x, y - std_dev, y + std_dev, color='fuchsia', alpha=0.2, label='Performance Std')

plt.xlabel('(d) Macro Search Epoch ' + r'$E_{mac}$', fontdict=font1)
plt.ylabel('Average Validation Accuracy', fontdict=font1)
plt.legend(loc=4, prop=font2)

# 显示图形
plt.show()
```

## 多均值折线图

```python
import numpy as np
import copy
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator

def _2d_line(x, y):

    """
    解决 Ubuntu　matplotlib Times New Roman 字体设置问题:
    $ sudo apt install msttcorefonts -qq
    $ rm ~/.cache/matplotlib -rf
    """
    # 输出图片大小设置
    # plt.figure(figsize=(7, 3.5))

    # 坐标轴，标题字体大小，字体粗细，字体类型设置
    font1 = {'family': 'Times New Roman',
             'weight': 'normal',
             'size': 15,
             }

    font2 = {'family': 'Times New Roman',
             'weight': 'normal',
             'size': 12,
             }

    plt.subplot(2, 2, 1)
    # 坐标轴标题，图片标题显示
    plt.xlabel("(a) Warm-up Epoch", fontdict=font1)
    plt.ylabel("Average Validation Performance", fontdict=font1)

    # 坐标轴间隔设置，ｘ间隔为100,y将为0.005
    x_axis_space = MultipleLocator(30)
    y_axis_space = MultipleLocator(0.01)
    # y_axis_space = MultipleLocator(0.005)

    # 获取坐标轴对象
    axis = plt.gca()
    # 使用坐标轴实例设置坐标轴间隔
    axis.xaxis.set_major_locator(x_axis_space)
    axis.yaxis.set_major_locator(y_axis_space)

    # 设置x轴刻度显示
    value = ["10", "30", "50", "70", "90", "110", "130", "150", "170"]
    plt.xticks(x[0], value)

    # 限制坐标轴显示范围, x[-10,2600],y[0.74,8.45]
    plt.xlim(5, 180)
    plt.ylim(0.89, 0.97) # CS Photo

    # 设置坐标轴刻度值大小以及字体
    plt.tick_params(labelsize=13)
    labels = axis.get_xticklabels() + axis.get_yticklabels()
    [label.set_fontname('Times New Roman') for label in labels]

    # 折线图作图
    x1 = x[0]
    y1 = y[0]

    x2 = x[1]
    y2 = y[1]

    l1 = plt.plot(x1, y1, "-o", color="lightskyblue", label="CS")
    l2 = plt.plot(x2, y2, "-v", color="lightcoral", label="Photo")
    plt.grid(linestyle="--")
    plt.legend(loc="best", prop=font2)

    # 图2
    plt.subplot(2, 2, 2)
    # 坐标轴标题，图片标题显示
    plt.xlabel("(b) Differentiable Search Size x100", fontdict=font1)
    plt.ylabel("Average Validation Performance", fontdict=font1)

    # 坐标轴间隔设置
    x_axis_space = MultipleLocator(100)
    y_axis_space = MultipleLocator(0.01)

    # 获取坐标轴对象
    axis = plt.gca()
    # 使用坐标轴实例设置坐标轴间隔
    axis.xaxis.set_major_locator(x_axis_space)
    axis.yaxis.set_major_locator(y_axis_space)

    # 设置x轴刻度显示
    value = ["1", "3", "5", "7", "9", "11", "13", "15", "17"]
    plt.xticks(x[2], value)

    # 限制坐标轴显示范围, x[-10,2600],y[0.74,8.45]
    plt.xlim(50, 1800)
    plt.ylim(0.90, 0.97)  # CS Photo

    # 设置坐标轴刻度值大小以及字体
    plt.tick_params(labelsize=13)
    labels = axis.get_xticklabels() + axis.get_yticklabels()
    [label.set_fontname('Times New Roman') for label in labels]

    # 折线图作图
    x3 = x[2]
    y3 = y[2]

    x4 = x[3]
    y4 = y[3]

    l3 = plt.plot(x3, y3, "-o", color="lightskyblue", label="CS")
    l4 = plt.plot(x4, y4, "-v", color="lightcoral", label="Photo")

    # 开启网格
    plt.grid(linestyle="--")
    plt.legend(loc="lower left", prop=font2)

    # 图3
    plt.subplot(2, 2, 3)
    # 坐标轴标题，图片标题显示
    plt.xlabel("(c) Temperature Coefficient", fontdict=font1)
    plt.ylabel("Average Validation Performance", fontdict=font1)

    # 坐标轴间隔设置
    x_axis_space = MultipleLocator(0.2)
    y_axis_space = MultipleLocator(0.01)

    # 获取坐标轴对象
    axis = plt.gca()
    # 使用坐标轴实例设置坐标轴间隔
    axis.xaxis.set_major_locator(x_axis_space)
    axis.yaxis.set_major_locator(y_axis_space)

    # 设置x轴刻度显示
    value = ["0.1", "0.3", "0.5", "0.7", "0.9", "1.1", "1.3", "1.5"]
    plt.xticks(x[4][1:], value)

    # 限制坐标轴显示范围, x[-10,2600],y[0.74,8.45]
    plt.xlim(0, 1.6)
    plt.ylim(0.93, 0.965)  # CS Photo

    # 设置坐标轴刻度值大小以及字体
    plt.tick_params(labelsize=13)
    labels = axis.get_xticklabels() + axis.get_yticklabels()
    [label.set_fontname('Times New Roman') for label in labels]

    # 折线图作图
    x5 = x[4][1:]
    y5 = y[4][1:]

    x6 = x[5][1:]
    y6 = y[5][1:]

    l5 = plt.plot(x5, y5, "-o", color="lightskyblue", label="CS")
    l6 = plt.plot(x6, y6, "-v", color="lightcoral", label="Photo")

    # 开启网格
    plt.grid(linestyle="--")
    plt.legend(loc="best",prop=font2)

    # 图4
    plt.subplot(2, 2, 4)
    # 坐标轴标题，图片标题显示
    plt.xlabel("(d) Single-path Training Epoch", fontdict=font1)
    plt.ylabel("Average Validation Performance", fontdict=font1)

    # 坐标轴间隔设置，ｘ间隔为100,y将为0.005
    x_axis_space = MultipleLocator(20)
    y_axis_space = MultipleLocator(0.01)
    # y_axis_space = MultipleLocator(0.005)

    # 获取坐标轴对象
    axis = plt.gca()
    # 使用坐标轴实例设置坐标轴间隔
    axis.xaxis.set_major_locator(x_axis_space)
    axis.yaxis.set_major_locator(y_axis_space)

    # 设置x轴刻度显示
    value = ["20", "40", "60", "80", "100", "120", "140", "160", "180"]
    plt.xticks(x[6], value)

    # 限制坐标轴显示范围, x[-10,2600],y[0.74,8.45]
    plt.xlim(10, 190)
    plt.ylim(0.91, 0.97)  # CS Photo

    # 设置坐标轴刻度值大小以及字体
    plt.tick_params(labelsize=13)
    labels = axis.get_xticklabels() + axis.get_yticklabels()
    [label.set_fontname('Times New Roman') for label in labels]

    # 折线图作图
    x7 = x[6]
    y7 = y[6]

    x8 = x[7]
    y8 = y[7]

    l7 = plt.plot(x7, y7, "-o", color="lightskyblue", label="CS")
    l8 = plt.plot(x8, y8, "-v", color="lightcoral", label="Photo")
    plt.grid(linestyle="--")
    plt.legend(loc="best",prop=font2)

    plt.show()

if __name__=="__main__":

    x_list = [[10.0, 30.0, 50.0, 70.0, 90.0, 110.0, 130.0, 150.0, 170.0],
              [10.0, 30.0, 50.0, 70.0, 90.0, 110.0, 130.0, 150.0, 170.0],
              [100.0, 300.0, 500.0, 700.0, 900.0, 1100.0, 1300.0, 1500.0, 1700.0],
              [100.0, 300.0, 500.0, 700.0, 900.0, 1100.0, 1300.0, 1500.0, 1700.0],
              [0.05, 0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3, 1.5],
              [0.05, 0.1, 0.3, 0.5, 0.7, 0.9, 1.1, 1.3, 1.5],
              [20.0, 40.0, 60.0, 80.0, 100.0, 120.0, 140.0, 160.0, 180.0],
              [20.0, 40.0, 60.0, 80.0, 100.0, 120.0, 140.0, 160.0, 180.0]]

    y_list = [[0.9282222222222222, 0.9304444444444446, 0.9237777777777778, 0.953111111111111, 0.9020000000000001, 0.9413333333333334, 0.9533333333333331, 0.9326666666666666, 0.9177777777777777],
              [0.9283333333333331, 0.9370833333333332, 0.9470833333333333, 0.9504166666666667, 0.9512500000000002, 0.9570833333333335, 0.9595833333333335, 0.9516666666666668, 0.9429166666666665],
              [0.9573333333333333, 0.9297777777777778, 0.9506666666666665, 0.9164444444444445, 0.9573333333333334, 0.9584444444444443, 0.9417777777777777, 0.9559999999999998, 0.906],
              [0.9379166666666665, 0.9520833333333334, 0.95, 0.9512500000000002, 0.9499999999999998, 0.9583333333333334, 0.95625, 0.94625, 0.9454166666666666],
              [0.9506666666666665, 0.9386666666666666, 0.9333333333333333, 0.9571111111111111, 0.9513333333333336, 0.9495555555555555, 0.944888888888889, 0.9486666666666667, 0.9466666666666667],
              [0.9441666666666666, 0.9558333333333332, 0.945, 0.9566666666666667, 0.9525, 0.9533333333333334, 0.9487500000000001, 0.9495833333333332, 0.9504166666666667],
              [0.9586666666666668, 0.9515555555555556, 0.9315555555555555, 0.9608888888888888, 0.9113333333333336, 0.9366666666666668, 0.9295555555555557, 0.9297777777777778, 0.9337777777777777],
              [0.9445833333333333, 0.9525, 0.9541666666666666, 0.9554166666666666, 0.9420833333333333, 0.9483333333333333, 0.95375, 0.9337500000000001, 0.9541666666666668]]


    x_list = np.array(x_list)
    y_list = np.array(y_list)
    _2d_line(x_list, y_list)
```
