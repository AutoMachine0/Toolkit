# 多变量柱状图

```python
import matplotlib.pyplot as plt
import numpy as np

def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha='center', va='bottom',
                    font=font1)

font1 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 15}

font2 = {'family': 'Times New Roman',
         'weight': 'normal',
         'size': 13}

labels = ['THYROID', 'SATELLITE']
LUNAR = [85.44, 85.35]
AdaGramp = [92.83, 79.53]

x = np.arange(len(labels))

# 调整柱状图宽度
width = 0.35

# 获取作图句柄
fig, ax = plt.subplots()

# 柱状图作图
rects1 = ax.bar(x - width/2, LUNAR, width, label='Distance-based (LUNAR)', color='orange')
rects2 = ax.bar(x + width/2, AdaGramp, width, label='Feature-based', color='deepskyblue')

# 设置x,y title内容及其字体配置
ax.set_ylabel('Average Test AUC Score', fontdict=font1)
ax.set_xlabel('(a) Performance Comparison between\nDistance-based and Feature-based Message Passing',
              fontdict=font1)

ax.set_xticks(x)
# 设置x轴显示内容与字体配置
ax.set_xticklabels(labels, font=font1)

# 限制柱状图y轴范围
plt.ylim(78, 94)
# 设置柱状图y轴字体配置
plt.yticks(font=font1)

# 增加柱状图数字
autolabel(rects1)
autolabel(rects2)

plt.grid(linestyle="--", axis="y")
ax.legend(loc=1, prop=font2)
fig.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
group_labels = ['ANNTHYROID', 'MI-F']  # 横坐标标签
categories = np.arange(9)

data1 = [85.44, 87.53, 87.90, 81.42, 84.81, 86.71, 90.96, 87.34, 90.53]  # 数据集 1 的值
data2 = [84.37, 82.64, 76.52, 83.37, 75.69, 75.06, 69.46, 62.27, 80.49]  # 数据集 2 的值

bar_width = 1.0
group_gap = 3.0
x1 = categories
x2 = categories + len(categories) + group_gap

colors = ['#6495ED', '#4169E1', '#48D1CC', '#20B2AA', '#fd8d49', '#ff5b00', "#D8BFD8", '#FFD700', "#DB7093"]

# 创建画布
plt.figure(figsize=(10, 6))

# 绘制柱状图并添加数值标签
for i in range(9):
    plt.bar(x1[i], data1[i], width=bar_width, color=colors[i], alpha=0.8)
    plt.text(x1[i], data1[i] + 0.5, f'{data1[i]:.2f}', ha='center', va='bottom', fontsize=13.5, fontname='Times New Roman')
    plt.bar(x2[i], data2[i], width=bar_width, color=colors[i], alpha=0.8)
    plt.text(x2[i], data2[i] + 0.5, f'{data2[i]:.2f}', ha='center', va='bottom', fontsize=13.5, fontname='Times New Roman')

# 设置 X 轴刻度和标签
xticks = [np.mean(x1), np.mean(x2)]
plt.xticks(xticks, group_labels, fontsize=15, fontname='Times New Roman')  # 横坐标字体设置

# 设置纵坐标刻度字体为 Times New Roman
plt.gca().yaxis.set_tick_params(labelsize=15)  # 设置刻度字体大小
plt.gca().tick_params(axis='y', which='major')  # 设置纵坐标主要刻度
for label in plt.gca().get_yticklabels():
    label.set_fontname("Times New Roman")  # 修改字体为 Times New Roman

# 添加标签和标题
plt.ylabel('Average Test AUC', fontsize=16, fontname='Times New Roman')  # 纵轴标签字体

# 设置图例
legend_labels = ["LUNAR-100", "LUNAR-20", "GCN-wo-edge", "GCN", "GAT-wo-edge", "GAT", "GADBench", "GIN", "GraphNAS"]
handles = [plt.Line2D([0], [0], color=colors[i], lw=5) for i in range(9)]
plt.legend(handles, legend_labels, loc='upper center', ncol=9, fontsize=11, prop={'family': 'Times New Roman'}, bbox_to_anchor=(0.5, 1.1))

# 显示网格
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 调整布局
plt.tight_layout()
plt.ylim(60, 94)

# 显示图表
plt.show()
```
