### 6.添加系统服务
1. 创建systemd服务单元文件，用于定义、管理clash服务的启动、停止和重启等行为

```python
sudo vim /etc/systemd/system/clash.service
```
2. 按i键进入文本编辑模式，复制配置内容到clash.service文件

```python
[Unit]
Description=Clash #网络代理守护进程
After=network-online.target

[Service]
Type=simple
Restart=always
ExecStart=/usr/local/clash/clash -d /usr/local/clash #/usr/local/clash/clash 是可执行程序clash的路径，-d 指代指定可执行程序clash所在的目录

[Install]
WantedBy=multi-user.target
```
3. shift+Esc退出编辑模式，输入:wq命令保存退出
4. 编辑完成，重载systemctl并设置开机启动

```python
sudo systemctl daemon-reload
sudo systemctl enable clash
```
5. 终端测试vpn配置是否成功

```python
curl https://www.google.com
```
6. 关闭vpn服务只需在Ubuntu系统中关闭代理即可
